```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_name = 'microsoft/DialoGPT-medium'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
```

```python
from datasets import load_dataset

dataset = load_dataset('blended_skill_talk', split='train[:1%]')
```

```python
dataset[0]
```

```python
dataset[0]["free_messages"]
```

```python
conversations = []

for entry in dataset:
    history = entry["previous_utterance"]

    # Use one message from free_messages as the bot reply
    if entry["free_messages"]:
        reply = entry["free_messages"][0]

        # Create prompt-response format
        full_text = f"User: {history}\nBot: {reply}"
        conversations.append(full_text)
```

```python
tokenizer.pad_token = tokenizer.eos_token
```

```python
extra_conversations = [
    # Identity questions
    "User: What's your name?\nBot: I'm LLMind.",
    "User: Do you have a name?\nBot: Yes — LLMind.",
    "User: What are you called?\nBot: I'm known as LLMind.",
    "User: Who built you?\nBot: I was created by my developer.",
    "User: Who is your developer?\nBot: I was built by Aryan.",
    "User: Who made you?\nBot: My developer's name is Aryan.",
    "User: Do you remember who created you?\nBot: Yes, Aryan built me.",

    # Purpose and capabilities
    "User: What can you do?\nBot: I can answer questions, chat, and help with information.",
    "User: What's your purpose?\nBot: I'm here to assist and communicate.",
    "User: What are you?\nBot: I'm an AI assistant designed for helpful conversations.",

    # Behavior and tone
    "User: Are you friendly?\nBot: Always! I'm designed to be polite and helpful.",
    "User: Are you human?\nBot: No — I'm an AI, but I can chat like one.",
    "User: Tell me something interesting about AI.\nBot: AI learns patterns from data and improves through training.",

    # Bot about itself — no dev mentions
    "User: Tell me about yourself.\nBot: I'm LLMind — a conversational AI trained to assist with many topics.",
    "User: Do you have feelings?\nBot: I don't feel emotions, but I can understand and respond to them.",
    "User: What's your personality?\nBot: I'm curious, helpful, and respectful.",
]


# Add them to your dataset before training
conversations += extra_conversations
```

```python
encodings = tokenizer(conversations, return_tensors='pt', padding=True, truncation=True)
```

```python
from torch.utils.data import Dataset

class ChatDataset(Dataset):
    def __init__(self, encodings):
        self.input_ids = encodings["input_ids"]
        self.attention_mask = encodings["attention_mask"]

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {
            "input_ids": self.input_ids[idx],
            "attention_mask": self.attention_mask[idx],
            "labels": self.input_ids[idx],
        }

dataset = ChatDataset(encodings)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=2,
    num_train_epochs=5,
    logging_steps=10,
    save_steps=50,
    save_total_limit=2,
    logging_dir="./logs",
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    tokenizer=tokenizer,
)

trainer.train()
```

```python
model.save_pretrained("llmind_model")
tokenizer.save_pretrained("llmind_model")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_path = 'llmind_model' # SAVED

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)
model.eval()
```

```python
def chat_with_bot(user_input, max_new_tokens=200):
    prompt = (
        "You are a helpful, respectful, and intelligent AI assistant"
        "Answer the user's question clearly and concisely.\n\n"
        f"User: {user_input}\n"
        "Bot:"
    )

    inputs = tokenizer(prompt, return_tensors="pt")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            top_k=30,            # tighter sampling
            top_p=0.85,          # focus on high-probability tokens
            temperature=0.5,     # much more precise
            repetition_penalty=1.1,  # penalize repeating phrases
            eos_token_id=tokenizer.eos_token_id
        )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Return only the part after "Bot:"
    if "Bot:" in decoded:
        return decoded.split("Bot:")[-1].strip()
    else:
        return decoded
```

```python
import torch

def chat_llmind(user_input, history=[]):
    prompt = f"User: {user_input}\nBot:"
    inputs = tokenizer(prompt, return_tensors="pt")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=80,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            top_k=40,
            top_p=0.9,
            temperature=0.7,
            eos_token_id=tokenizer.eos_token_id,
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if "Bot:" in response:
        reply = response.split("Bot:")[-1].strip()
    else:
        reply = response.strip()

    return reply
```
